# 6. Préparer le modèle et le tokenizer

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import prepare_model_for_kbit_training, LoraConfig, get_peft_model
import torch

# ✅ Modèle choisi
model_name = "bigscience/bloomz-3b"

# Charger le tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Vérifier le pad_token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Config quantization 4-bit
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

# Charger le modèle causal LM
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto"
)

# Préparer le modèle pour LoRA
model = prepare_model_for_kbit_training(model)

# Config LoRA (pour BLOOM et BLOOMZ, les bons modules sont "query_key_value")
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["query_key_value"],
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"
)

# Appliquer LoRA
model = get_peft_model(model, lora_config)

# Afficher les paramètres entraînables
model.print_trainable_parameters()
```

# Tokenisation des données

```python
# ✅ Ajouter un pad_token au tokenizer si non défini
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})

# ✅ Redimensionner les embeddings du modèle si tu ajoutes le token (optionnel ici si tu ne réutilises pas le modèle maintenant)
# model.resize_token_embeddings(len(tokenizer))  # à activer si tu charges le modèle ensuite

# ✅ Fonction de tokenization corrigée
def tokenize(example):
    return tokenizer(example["text"], truncation=True, padding="max_length", max_length=512)

# ✅ Application à l'ensemble du dataset
tokenized_dataset = dataset.map(tokenize, batched=True, remove_columns=["text"])
```

# 9. Configurer les arguments d’entraînement

```python
from transformers import TrainingArguments
from trl import SFTTrainer

training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/bigscience_arabicoptimiae",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    warmup_steps=10,
    max_steps=2000,
    learning_rate=2e-4,
    logging_steps=10,
    optim="adamw_torch",
    save_steps=100,
    save_total_limit=2,
    fp16=True,
    report_to="none"

)
```

# 10. Lancer l’entraînement

```python
from trl import SFTTrainer

# Assure-toi que `tokenized_dataset` a bien été préparé avant ce bloc !

trainer = SFTTrainer(
    model=model,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    peft_config=lora_config,
    args=training_args
)

# ✅ Lancer l’entraînement
trainer.train()

```

```python
import json

jsonl_path = "/content/drive/MyDrive/prepared_arabic_qa_generation.jsonl"
prompts, true_answers = [], []

with open(jsonl_path, "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        if "### الإجابة:" in obj["text"]:
            q, a = obj["text"].split("### الإجابة:")
            prompts.append(q.strip() + "### الإجابة:\n")
            true_answers.append(a.strip())
        if len(prompts) >= 100:
            break
```

```python
from tqdm import tqdm
import numpy as np
from sklearn.metrics import f1_score, accuracy_score
from bert_score import score as bert_score
from rouge_score import rouge_scorer

def evaluate_rag_pipeline(pipe, questions, references, max_new_tokens=50):
    predictions = []

    print("🔁 Génération des réponses...")
    for q in tqdm(questions):
        output = pipe(q, max_new_tokens=max_new_tokens, do_sample=False, temperature=0.0)
        generated = output[0]["generated_text"]
        answer = generated.split("الإجابة هي:")[-1].strip().split("\n")[0].strip(" .،")
        predictions.append(answer)

    # === 1. Exact Match
    exact_matches = [int(p.strip() == r.strip()) for p, r in zip(predictions, references)]
    em_score = np.mean(exact_matches)

    # === 2. Word-level F1 Score
    def f1_individual(pred, ref):
        pred_tokens = set(pred.split())
        ref_tokens = set(ref.split())
        common = pred_tokens & ref_tokens
        if not common:
            return 0
        precision = len(common) / len(pred_tokens)
        recall = len(common) / len(ref_tokens)
        return 2 * precision * recall / (precision + recall)

    f1_scores = [f1_individual(p, r) for p, r in zip(predictions, references)]

    # === 3. ROUGE-L
    rouge = rouge_scorer.RougeScorer(["rougeL"], use_stemmer=True)
    rouge_scores = [rouge.score(p, r)["rougeL"].fmeasure for p, r in zip(predictions, references)]

    # === 4. BERTScore (lang="ar" pour l’arabe)
    P, R, F1 = bert_score(predictions, references, lang="ar", rescale_with_baseline=True)

    # === 5. Résumé des résultats
    print("\n📊 Évaluation de la génération :")
    print(f"🔹 Exact Match: {em_score:.4f}")
    print(f"🔹 Word-level F1: {np.mean(f1_scores):.4f}")
    print(f"🔹 ROUGE-L: {np.mean(rouge_scores):.4f}")
    print(f"🔹 BERTScore-F1: {F1.mean().item():.4f}")

    return {
        "exact_match": em_score,
        "word_f1": np.mean(f1_scores),
        "rougeL": np.mean(rouge_scores),
        "bertscore_f1": F1.mean().item()
    }
```

```python
import json
import numpy as np
from tqdm import tqdm
from rank_bm25 import BM25Okapi
from sklearn.preprocessing import normalize
from sentence_transformers import SentenceTransformer
from rouge_score import rouge_scorer
from bert_score import score as bert_score
import faiss
import pandas as pd
from IPython.display import display

# === 1. Charger les données ===
file_path = "/content/drive/MyDrive/prepared_arabic_qa_generation.jsonl"
texts, tokenized_texts, questions, answers = [], [], [], []

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        text = obj["text"]
        if "### سؤال:" in text and "### الإجابة:" in text:
            q, a = text.split("### الإجابة:")
            q = q.strip().replace("### سؤال:", "").strip()
            a = a.strip()
            questions.append(q)
            answers.append(a)
            texts.append(q)
            tokenized_texts.append(q.lower().split())
        if len(questions) >= 50:  # Réduction pour faible ressource
            break

# === 2. Index BM25 ===
bm25 = BM25Okapi(tokenized_texts)

# === 3. Embeddings distiluse + FAISS ===
dense_model = SentenceTransformer("sentence-transformers/distiluse-base-multilingual-cased-v2", device="cpu")
dense_embeddings = dense_model.encode(texts, convert_to_numpy=True, show_progress_bar=True, batch_size=8)
dense_embeddings = normalize(dense_embeddings)

dimension = dense_embeddings.shape[1]
faiss_index = faiss.IndexFlatIP(dimension)
faiss_index.add(dense_embeddings)

# === 4. Fusion Hybride ===
def hybrid_retrieve(query, top_k=5, alpha=0.5):
    bm25_scores = bm25.get_scores(query.lower().split())
    query_embedding = dense_model.encode([query], convert_to_numpy=True)
    query_embedding = normalize(query_embedding)
    dense_scores, _ = faiss_index.search(query_embedding, len(texts))
    dense_scores = dense_scores.flatten()
    hybrid_scores = alpha * bm25_scores + (1 - alpha) * dense_scores
    top_indices = np.argsort(hybrid_scores)[::-1][:top_k]
    return [texts[i] for i in top_indices], [answers[i] for i in top_indices], top_indices

# === 5. Évaluation ===
rouge = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
em_list, f1_list, rouge_list, mrr_list, recall_list = [], [], [], [], []
predictions, references = [], []

for i in tqdm(range(len(questions))):
    q = questions[i]
    true_answer = answers[i]

    _, top_answers, _ = hybrid_retrieve(q, top_k=5)
    predictions.append(top_answers[0])
    references.append(true_answer)

    # Exact Match
    em_list.append(int(top_answers[0].strip() == true_answer.strip()))

    # Word-level F1
    pred_tokens = top_answers[0].split()
    ref_tokens = true_answer.split()
    common = set(pred_tokens) & set(ref_tokens)
    if not common:
        f1 = 0.0
    else:
        p = len(common) / len(pred_tokens)
        r = len(common) / len(ref_tokens)
        f1 = 2 * p * r / (p + r)
    f1_list.append(f1)

    # ROUGE-L
    rouge_l = rouge.score(top_answers[0], true_answer)['rougeL'].fmeasure
    rouge_list.append(rouge_l)

    # Recall@5 et MRR@5
    found = False
    for rank, ans in enumerate(top_answers):
        if ans.strip() == true_answer.strip():
            recall_list.append(1)
            mrr_list.append(1 / (rank + 1))
            found = True
            break
    if not found:
        recall_list.append(0)
        mrr_list.append(0)

# === 6. BERTScore ===
P, R, F1_bert = bert_score(predictions, references, lang="ar", model_type="bert-base-multilingual-cased")

# === 7. Résumé final ===
results = {
    "Recall@5": round(np.mean(recall_list), 4),
    "MRR@5": round(np.mean(mrr_list), 4),
    "Exact Match (EM)": round(np.mean(em_list), 4),
    "Word-level F1": round(np.mean(f1_list), 4),
    "ROUGE-L": round(np.mean(rouge_list), 4),
    "BERTScore-F1": round(F1_bert.mean().item(), 4)
}

df = pd.DataFrame([results])
display(df)
```

```python
import streamlit as st
import json
import numpy as np
from rank_bm25 import BM25Okapi
from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import normalize
import faiss
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM

# === 1. Charger les données ===
@st.cache_data
def load_data(jsonl_path):
    texts, tokenized_texts = [], []
    with open(jsonl_path, "r", encoding="utf-8") as f:
        for line in f:
            obj = json.loads(line)
            text = obj["text"]
            if "### سؤال:" in text and "### الإجابة:" in text:
                q, a = text.split("### الإجابة:")
                q = q.strip().replace("### سؤال:", "").strip()
                texts.append(f"{q}\n### الإجابة:{a.strip()}")
                tokenized_texts.append(q.lower().split())
    return texts, tokenized_texts

# === 2. Index BM25 + FAISS ===
@st.cache_resource
def build_indices(texts, tokenized_texts):
    bm25 = BM25Okapi(tokenized_texts)
    model = SentenceTransformer("CAMeL-Lab/bert-base-arabic-camelbert-mix")
    embeddings = model.encode(texts, convert_to_numpy=True, show_progress_bar=True)
    embeddings = normalize(embeddings)
    index = faiss.IndexFlatIP(embeddings.shape[1])
    index.add(embeddings)
    return bm25, model, embeddings, index

# === 3. Récupération Hybride ===
def hybrid_retrieve(query, bm25, model, faiss_index, texts, top_k=3, alpha=0.5):
    bm25_scores = bm25.get_scores(query.lower().split())
    query_embedding = normalize(model.encode([query], convert_to_numpy=True))
    dense_scores, _ = faiss_index.search(query_embedding, len(texts))
    dense_scores = dense_scores.flatten()
    combined_scores = alpha * bm25_scores + (1 - alpha) * dense_scores
    top_indices = np.argsort(combined_scores)[::-1][:top_k]
    return [texts[i] for i in top_indices]

# === 4. Générateur (GPT-Neo ou GPT2 fine-tuné) ===
@st.cache_resource
def load_generator(model_path):
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    model = AutoModelForCausalLM.from_pretrained(model_path)
    tokenizer.pad_token = tokenizer.eos_token  # obligatoire pour GPT2/GPTNeo
    return pipeline("text-generation", model=model, tokenizer=tokenizer, device=0)

# === 5. Interface Streamlit ===
st.title("🤖 Générateur Arabe QA avec RAG Hybride")
st.markdown("Posez une question, et le système récupère les documents et génère une réponse !")

# Fichier et modèle
json_path = "/content/drive/MyDrive/prepared_arabic_qa_generation.jsonl"
model_path = "/content/drive/MyDrive/final_model_gptneo_arabic_gen"

texts, tokenized_texts = load_data(json_path)
bm25, retriever_model, dense_embeddings, faiss_index = build_indices(texts, tokenized_texts)
generator = load_generator(model_path)

# Entrée utilisateur
query = st.text_input("📝 Votre question :", "كم عدد سور القرآن الكريم؟")

if st.button("📥 Générer une réponse"):
    retrieved = hybrid_retrieve(query, bm25, retriever_model, faiss_index, texts)
    context = "\n".join(retrieved)

    # Créer le prompt structuré
    prompt = f"### سؤال:\n{query}\n\n### سياق:\n{context}\n\n### الإجابة:"

    # Générer la réponse
    generated = generator(prompt, max_new_tokens=100, do_sample=True, temperature=0.7)[0]['generated_text']

    # Extraire uniquement la partie après ### الإجابة:
    if "### الإجابة:" in generated:
        final_answer = generated.split("### الإجابة:")[-1].strip()
    else:
        final_answer = generated.strip()

    # Affichage
    st.subheader("📚 Contexte Récupéré")
    st.code(context, language='markdown')

    st.subheader("💬 Réponse Générée")
    st.success(final_answer)
```

```python
def hybrid_retrieve(query, top_k=3, alpha=0.5):
    # Sparse scores
    bm25_scores = bm25.get_scores(query.lower().split())
    bm25_top = np.argsort(bm25_scores)[::-1][:top_k]

    # Dense scores
    dense_vec = embedder.encode([query], convert_to_numpy=True).astype("float32")
    faiss_scores, faiss_top = index.search(dense_vec, top_k)

    # Combine scores
    hybrid_scores = {}
    for i in range(top_k):
        hybrid_scores[bm25_top[i]] = alpha * bm25_scores[bm25_top[i]]
        hybrid_scores[faiss_top[0][i]] = hybrid_scores.get(faiss_top[0][i], 0) + (1 - alpha) * faiss_scores[0][i]

    ranked = sorted(hybrid_scores.items(), key=lambda x: -x[1])
    return [docs[idx] for idx, _ in ranked[:1]]  # top 1 context
```

```python
from rouge_score import rouge_scorer
from bert_score import score as bert_score
import numpy as np

def evaluate_rag_hybrid(pipe, questions, references, max_new_tokens=50):
    predictions = []

    print("🔁 Génération avec récupération hybride...")
    for q in tqdm(questions):
        context = hybrid_retrieve(q)[0]
        prompt = f"### سؤال: {q}\n{context}\n### الإجابة:"
        output = pipe(prompt, max_new_tokens=max_new_tokens, do_sample=False, temperature=0.0)
        answer = output[0]["generated_text"].split("### الإجابة:")[-1].strip().split("\n")[0].strip(" .،")
        predictions.append(answer)

    # Exact Match
    em_score = np.mean([int(p.strip() == r.strip()) for p, r in zip(predictions, references)])

    # Word-level F1
    def f1_individual(pred, ref):
        pred_tokens = set(pred.split())
        ref_tokens = set(ref.split())
        common = pred_tokens & ref_tokens
        if not common:
            return 0
        precision = len(common) / len(pred_tokens)
        recall = len(common) / len(ref_tokens)
        return 2 * precision * recall / (precision + recall)
    f1_scores = [f1_individual(p, r) for p, r in zip(predictions, references)]

    # ROUGE-L
    rouge = rouge_scorer.RougeScorer(["rougeL"], use_stemmer=True)
    rouge_scores = [rouge.score(p, r)["rougeL"].fmeasure for p, r in zip(predictions, references)]

    # BERTScore
    P, R, F1 = bert_score(predictions, references, lang="ar", rescale_with_baseline=True)

    # Résultats
    print("\n📊 Évaluation de la génération avec RAG hybride :")
    print(f"🔹 Exact Match: {em_score:.4f}")
    print(f"🔹 Word-level F1: {np.mean(f1_scores):.4f}")
    print(f"🔹 ROUGE-L: {np.mean(rouge_scores):.4f}")
    print(f"🔹 BERTScore-F1: {F1.mean().item():.4f}")

    return {
        "exact_match": em_score,
        "word_f1": np.mean(f1_scores),
        "rougeL": np.mean(rouge_scores),
        "bertscore_f1": F1.mean().item()
    }
```

```python
import json
import numpy as np
from tqdm import tqdm
from rank_bm25 import BM25Okapi
from sklearn.preprocessing import normalize
from sentence_transformers import SentenceTransformer
from rouge_score import rouge_scorer
from bert_score import score as bert_score
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import faiss
import pandas as pd

# === 1. Charger les données ===
file_path = "/content/drive/MyDrive/prepared_arabic_qa_generation.jsonl"
texts, tokenized_texts, questions, answers = [], [], [], []

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        text = obj["text"]
        if "### سؤال:" in text and "### الإجابة:" in text:
            q, a = text.split("### الإجابة:")
            q = q.strip().replace("### سؤال:", "").strip()
            a = a.strip()
            questions.append(q)
            answers.append(a)
            texts.append(q)
            tokenized_texts.append(q.lower().split())
        if len(questions) >= 50:
            break

# === 2. Index BM25 ===
bm25 = BM25Okapi(tokenized_texts)

# === 3. Dense embeddings + FAISS ===
dense_model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2", device="cpu")
dense_embeddings = dense_model.encode(texts, convert_to_numpy=True, show_progress_bar=True, batch_size=8)
dense_embeddings = normalize(dense_embeddings)
faiss_index = faiss.IndexFlatIP(dense_embeddings.shape[1])
faiss_index.add(dense_embeddings)

# === 4. Hybride Retriever ===
def hybrid_retrieve(query, top_k=5, alpha=0.5):
    bm25_scores = bm25.get_scores(query.lower().split())
    query_emb = dense_model.encode([query], convert_to_numpy=True)
    query_emb = normalize(query_emb)
    dense_scores, _ = faiss_index.search(query_emb, len(texts))
    dense_scores = dense_scores.flatten()
    hybrid_scores = alpha * bm25_scores + (1 - alpha) * dense_scores
    top_indices = np.argsort(hybrid_scores)[::-1][:top_k]
    return [texts[i] for i in top_indices], [answers[i] for i in top_indices], top_indices

# === 5. Génération + Évaluation ===
models = {
    "distilgpt2": "/content/drive/MyDrive/googlesmall",
    "gptneox": "/content/drive/MyDrive/final_model_gptneo_arabic_gen",
    "aragpt2-medium": "/content/drive/MyDrive/merged_aragpt2-medium_arabic",

}

rouge = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
results = []

for name, path in models.items():
    print(f"\n🚀 Évaluation du modèle: {name}")
    tokenizer = AutoTokenizer.from_pretrained(path)
    model = AutoModelForCausalLM.from_pretrained(path)
    pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, device_map="auto")

    predicted_answers = []
    retrieved_answers = []
    em_list, f1_list, rouge_list, mrr_list, recall_list = [], [], [], [], []

    for i in tqdm(range(len(questions))):
        q = questions[i]
        true_answer = answers[i]

        top_texts, top_answers, _ = hybrid_retrieve(q)
        context = top_texts[0]
        prompt = f"{q}\n{context}\n### الإجابة:\n"

        gen_output = pipe(prompt, max_new_tokens=50, do_sample=False, temperature=0.0)
        generated = gen_output[0]["generated_text"]
        gen_answer = generated.split("### الإجابة:")[-1].strip().split("\n")[0].strip(" .،")

        predicted_answers.append(gen_answer)
        retrieved_answers.append(top_answers[0])

        # === Évaluation récupération (top-1) ===
        em_list.append(int(top_answers[0].strip() == true_answer.strip()))
        pred_tokens = top_answers[0].split()
        ref_tokens = true_answer.split()
        common = set(pred_tokens) & set(ref_tokens)
        f1 = 0 if not common else 2 * len(common) / (len(pred_tokens) + len(ref_tokens))
        f1_list.append(f1)
        rouge_l = rouge.score(top_answers[0], true_answer)['rougeL'].fmeasure
        rouge_list.append(rouge_l)
        found = False
        for rank, ans in enumerate(top_answers):
            if ans.strip() == true_answer.strip():
                recall_list.append(1)
                mrr_list.append(1 / (rank + 1))
                found = True
                break
        if not found:
            recall_list.append(0)
            mrr_list.append(0)

    # === BERTScore (génération) ===
    P, R, F1_bert = bert_score(predicted_answers, answers, lang="ar", model_type="bert-base-multilingual-cased")

    # === Génération - Exact Match, F1, ROUGE ===
    exact_matches = [int(p.strip() == a.strip()) for p, a in zip(predicted_answers, answers)]
    f1_scores = []
    rouge_scores = []

    for p, a in zip(predicted_answers, answers):
        pred_tokens = p.split()
        ref_tokens = a.split()
        common = set(pred_tokens) & set(ref_tokens)
        if not common:
            f1 = 0.0
        else:
            prec = len(common) / len(pred_tokens)
            rec = len(common) / len(ref_tokens)
            f1 = 2 * prec * rec / (prec + rec)
        f1_scores.append(f1)
        rouge_scores.append(rouge.score(p, a)['rougeL'].fmeasure)

    results.append({
        "Model": name,
        "Recall@5": round(np.mean(recall_list), 4),
        "MRR@5": round(np.mean(mrr_list), 4),
        "Retrieval EM": round(np.mean(em_list), 4),
        "Retrieval F1": round(np.mean(f1_list), 4),
        "Retrieval ROUGE": round(np.mean(rouge_list), 4),
        "Gen EM": round(np.mean(exact_matches), 4),
        "Gen F1": round(np.mean(f1_scores), 4),
        "Gen ROUGE": round(np.mean(rouge_scores), 4),
        "BERTScore-F1": round(F1_bert.mean().item(), 4),
    })

# === Affichage Résultats ===
df_results = pd.DataFrame(results)
print("\n📊 Tableau comparatif des modèles :")
display(df_results)
```
