# 6. PrÃ©parer le modÃ¨le et le tokenizer

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import prepare_model_for_kbit_training, LoraConfig, get_peft_model
import torch

# âœ… ModÃ¨le choisi
model_name = "bigscience/bloomz-3b"

# Charger le tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# VÃ©rifier le pad_token
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Config quantization 4-bit
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

# Charger le modÃ¨le causal LM
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto"
)

# PrÃ©parer le modÃ¨le pour LoRA
model = prepare_model_for_kbit_training(model)

# Config LoRA (pour BLOOM et BLOOMZ, les bons modules sont "query_key_value")
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["query_key_value"],
    lora_dropout=0.1,
    bias="none",
    task_type="CAUSAL_LM"
)

# Appliquer LoRA
model = get_peft_model(model, lora_config)

# Afficher les paramÃ¨tres entraÃ®nables
model.print_trainable_parameters()
```

# Tokenisation des donnÃ©es

```python
# âœ… Ajouter un pad_token au tokenizer si non dÃ©fini
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})

# âœ… Redimensionner les embeddings du modÃ¨le si tu ajoutes le token (optionnel ici si tu ne rÃ©utilises pas le modÃ¨le maintenant)
# model.resize_token_embeddings(len(tokenizer))  # Ã  activer si tu charges le modÃ¨le ensuite

# âœ… Fonction de tokenization corrigÃ©e
def tokenize(example):
    return tokenizer(example["text"], truncation=True, padding="max_length", max_length=512)

# âœ… Application Ã  l'ensemble du dataset
tokenized_dataset = dataset.map(tokenize, batched=True, remove_columns=["text"])
```

# 9. Configurer les arguments dâ€™entraÃ®nement

```python
from transformers import TrainingArguments
from trl import SFTTrainer

training_args = TrainingArguments(
    output_dir="/content/drive/MyDrive/bigscience_arabicoptimiae",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    warmup_steps=10,
    max_steps=2000,
    learning_rate=2e-4,
    logging_steps=10,
    optim="adamw_torch",
    save_steps=100,
    save_total_limit=2,
    fp16=True,
    report_to="none"

)
```

# 10. Lancer lâ€™entraÃ®nement

```python
from trl import SFTTrainer

# Assure-toi que `tokenized_dataset` a bien Ã©tÃ© prÃ©parÃ© avant ce bloc !

trainer = SFTTrainer(
    model=model,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    peft_config=lora_config,
    args=training_args
)

# âœ… Lancer lâ€™entraÃ®nement
trainer.train()

```

```python
from tqdm import tqdm
import numpy as np
from sklearn.metrics import f1_score, accuracy_score
from bert_score import score as bert_score
from rouge_score import rouge_scorer

def evaluate_rag_pipeline(pipe, questions, references, max_new_tokens=50):
    predictions = []

    print("ğŸ” GÃ©nÃ©ration des rÃ©ponses...")
    for q in tqdm(questions):
        output = pipe(q, max_new_tokens=max_new_tokens, do_sample=False, temperature=0.0)
        generated = output[0]["generated_text"]
        answer = generated.split("Ø§Ù„Ø¥Ø¬Ø§Ø¨Ø© Ù‡ÙŠ:")[-1].strip().split("\n")[0].strip(" .ØŒ")
        predictions.append(answer)

    # === 1. Exact Match
    exact_matches = [int(p.strip() == r.strip()) for p, r in zip(predictions, references)]
    em_score = np.mean(exact_matches)

    # === 2. Word-level F1 Score
    def f1_individual(pred, ref):
        pred_tokens = set(pred.split())
        ref_tokens = set(ref.split())
        common = pred_tokens & ref_tokens
        if not common:
            return 0
        precision = len(common) / len(pred_tokens)
        recall = len(common) / len(ref_tokens)
        return 2 * precision * recall / (precision + recall)

    f1_scores = [f1_individual(p, r) for p, r in zip(predictions, references)]

    # === 3. ROUGE-L
    rouge = rouge_scorer.RougeScorer(["rougeL"], use_stemmer=True)
    rouge_scores = [rouge.score(p, r)["rougeL"].fmeasure for p, r in zip(predictions, references)]

    # === 4. BERTScore (lang="ar" pour lâ€™arabe)
    P, R, F1 = bert_score(predictions, references, lang="ar", rescale_with_baseline=True)

    # === 5. RÃ©sumÃ© des rÃ©sultats
    print("\nğŸ“Š Ã‰valuation de la gÃ©nÃ©ration :")
    print(f"ğŸ”¹ Exact Match: {em_score:.4f}")
    print(f"ğŸ”¹ Word-level F1: {np.mean(f1_scores):.4f}")
    print(f"ğŸ”¹ ROUGE-L: {np.mean(rouge_scores):.4f}")
    print(f"ğŸ”¹ BERTScore-F1: {F1.mean().item():.4f}")

    return {
        "exact_match": em_score,
        "word_f1": np.mean(f1_scores),
        "rougeL": np.mean(rouge_scores),
        "bertscore_f1": F1.mean().item()
    }
```

```python
import json
import numpy as np
from tqdm import tqdm
from rank_bm25 import BM25Okapi
from sklearn.metrics import precision_recall_fscore_support
from sklearn.preprocessing import normalize
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from rouge_score import rouge_scorer
from bert_score import score as bert_score
import faiss

# === 1. Charger les donnÃ©es ===
file_path = "/content/drive/MyDrive/prepared_arabic_qa_generation.jsonl"
texts, tokenized_texts, questions, answers = [], [], [], []

with open(file_path, "r", encoding="utf-8") as f:
    for line in f:
        obj = json.loads(line)
        text = obj["text"]
        if "### Ø³Ø¤Ø§Ù„:" in text and "### Ø§Ù„Ø¥Ø¬Ø§Ø¨Ø©:" in text:
            q, a = text.split("### Ø§Ù„Ø¥Ø¬Ø§Ø¨Ø©:")
            q = q.strip().replace("### Ø³Ø¤Ø§Ù„:", "").strip()
            a = a.strip()
            questions.append(q)
            answers.append(a)
            texts.append(q)
            tokenized_texts.append(q.lower().split())

# === 2. BM25 Index ===
bm25 = BM25Okapi(tokenized_texts)

# === 3. Embeddings CAMeL + FAISS ===
dense_model = SentenceTransformer("CAMeL-Lab/bert-base-arabic-camelbert-mix")
dense_embeddings = dense_model.encode(texts, convert_to_numpy=True, show_progress_bar=True)
dense_embeddings = normalize(dense_embeddings)

dimension = dense_embeddings.shape[1]
faiss_index = faiss.IndexFlatIP(dimension)
faiss_index.add(dense_embeddings)

# === 4. Fusion Hybride + Reranking ===
def hybrid_retrieve(query, top_k=3, alpha=0.5):
    bm25_scores = bm25.get_scores(query.lower().split())
    query_embedding = dense_model.encode([query], convert_to_numpy=True)
    query_embedding = normalize(query_embedding)
    dense_scores, _ = faiss_index.search(query_embedding, len(texts))
    dense_scores = dense_scores.flatten()
    hybrid_scores = alpha * bm25_scores + (1 - alpha) * dense_scores
    top_indices = np.argsort(hybrid_scores)[::-1][:top_k]
    return [texts[i] for i in top_indices], [answers[i] for i in top_indices]

# === 5. Ã‰valuation ===
rouge = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
predictions = []
references = []

for q, ref in tqdm(zip(questions[:100], answers[:100]), total=100):
    _, top_answers = hybrid_retrieve(q, top_k=3, alpha=0.5)
    pred = top_answers[0]
    predictions.append(pred)
    references.append(ref)

# Exact Match
def exact_match(a, b):
    return int(a.strip() == b.strip())

exact_matches = [exact_match(p, r) for p, r in zip(predictions, references)]

# F1 word-level
def word_level_f1(pred, ref):
    pred_tokens = pred.split()
    ref_tokens = ref.split()
    common = set(pred_tokens) & set(ref_tokens)
    if not common:
        return 0.0
    precision = len(common) / len(pred_tokens)
    recall = len(common) / len(ref_tokens)
    return 2 * precision * recall / (precision + recall)

f1_scores = [word_level_f1(p, r) for p, r in zip(predictions, references)]

# ROUGE-L
rouge_scores = [rouge.score(p, r)['rougeL'].fmeasure for p, r in zip(predictions, references)]

# BERTScore (multilingual)
P, R, F1_bert = bert_score(predictions, references, lang="ar", model_type="bert-base-multilingual-cased")

# === 6. RÃ©sultats globaux ===
print(f"Exact Match (EM): {np.mean(exact_matches):.4f}")
print(f"Word-level F1: {np.mean(f1_scores):.4f}")
print(f"ROUGE-L: {np.mean(rouge_scores):.4f}")
print(f"BERTScore-F1: {F1_bert.mean().item():.4f}")
```
